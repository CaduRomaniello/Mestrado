```python
import pandas as pd
import json
from IPython.display import display

# Caminho para o arquivo logs.csv
file_path = "../output/logs.csv"

# Ler o arquivo CSV
try:
    logs_df = pd.read_csv(file_path)
    print("Arquivo lido com sucesso!")
    display(logs_df.head(1))  # Exibir as primeiras linhas do DataFrame
except FileNotFoundError:
    print(f"Erro: O arquivo '{file_path}' n√£o foi encontrado.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo: {e}")
```

```python
logs_df['accuracy'] = logs_df['accuracy'].round(2)
```

```python
grouped_df = logs_df.sort_values(by='accuracy', ascending=False).groupby('model_name')
    
# Exibir os primeiros resultados de cada grupo
for model, group in grouped_df:
    print(f"Modelo: {model}")
    display(group.head(1))  # Exibe o melhor resultado para cada modelo
    print(group.iloc[0]['config'])
```
